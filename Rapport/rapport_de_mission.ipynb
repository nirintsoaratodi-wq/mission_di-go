### **I. Comprendre les variables**


```python
#==========================
# Assignation de variables
#=========================
age = 20
nom = "Njaka"
#========================
# Affichager des variables
#========================
print(age)
print(nom)
#=======================================
# Double assignation sur une seule ligne
#=======================================
x, y = 10, 3
print(x)
print(y)
#-----------------------------------------
matiere , note= "math", 12
#note math est 12
print("note ", matiere ,"est" ,note )
#====================
# Types des variables
#====================
entier = 10
reel = 3.5
texte = '4'
booleen = False
is_female=True
print(type(entier))
print(type(reel)) # nombre avce virgule
print(type(texte)) # chaine de carractere 
print(type(booleen)) # Vraie ou Faux
```

### **II. Condition if\elif\else**

```python
note = 16
if note<10:
    print('Non admin')
elif 10<=note<12:
    print('Passable')
elif 12<=note<14:
    print('Assez-bien')
elif 14<=note<16:
    print('Bien')
else:
    print('Tres Bien')
```

### **III. boucle for** 

```python
for i in range(6):
    print(i)
```
