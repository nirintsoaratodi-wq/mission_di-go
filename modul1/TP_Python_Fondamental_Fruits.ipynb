# TP Python Fondamental (Data Science)

Manipulation simple des ventes de fruits.

````python
# Données de base
fruits = ["pomme", "mangue", "banane", "orange"]
prix = [800, 1500, 500, 600]  # Ariary / unité
ventes = [20, 12, 32, 25]     # Quantité vendue
````

## Exercice 1 : Variables
Calculer la recette totale de **bananes** et l'afficher.

````python
# Recette des bananes
recette_banane = prix[2] * ventes[2]
recette_banane
````

## Exercice 2 : Conditions
Afficher `Vente faible` si quantité < 15, sinon `Vente forte` pour chaque fruit.

````python
for i in range(len(fruits)):
    if ventes[i] < 15:
        print(fruits[i], ": Vente faible")
    else:
        print(fruits[i], ": Vente forte")
````

## Exercice 3 : Dictionnaire
Créer un dictionnaire regroupant les informations de vente.

````python
supermarche = {}
for i in range(len(fruits)):
    supermarche[fruits[i]] = {"prix": prix[i], "vente": ventes[i]}

supermarche
````

### Trouver le fruit le plus vendu

````python
max_vente = max(ventes)
fruit_max = fruits[ventes.index(max_vente)]
fruit_max
````

## Exercice 4 : Boucle
Calculer le bénéfice total.

````python
total = 0
for i in range(len(fruits)):
    total += prix[i] * ventes[i]
total
````

## Exercice 5 : NumPy
Calculer la recette totale avec NumPy.

````python
import numpy as np
prix_np = np.array(prix)
ventes_np = np.array(ventes)

np.sum(prix_np * ventes_np)
````

## Exercice 8 : Classe

````python
class Fruit:
    def __init__(self, nom, prix, vente):
        self.nom = nom
        self.prix = prix
        self.vente = vente
    def recette(self):
        return self.prix * self.vente

mangue = Fruit("mangue", 1500, 12)
mangue.recette()
````
